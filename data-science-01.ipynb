Ce que je veux afficher : Un histogramme du nombre de passage dans la rue du Gros Horloge

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('rouen_flux-pietons_vd.csv', delimiter=";")
```

Filtre le data_frame pour récupérer que les colonnes "RUE_DU_GROS_HORLOGE"

```python
gros_horloge_data = data_frame[data_frame['NOM_CAPTEUR'] == 'RUE_DU_GROS_HORLOGE']
```

Création de l'histogramme

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=gros_horloge_data, x='NOMBRE_PASSAGES', y='DATE_HAURE')
plt.title('Nombre de passages dans la rue du Gros Horloge')
plt.xlabel('Nombre de passages')
plt.ylabel('Date')
plt.grid(True)
```

Affichage de l'histogramme

```python
plt.show()
```
